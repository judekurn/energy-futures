# Pathways Diagrams

This notebook contains codes for generating pathways diagrams used in the documents called "Pathways Documents" to be sent to expert participants.

Updates history:
* 17 July 2018 - Beautify codes using standard Python naming convention (e.g. CAPS for constant etc.)
* 19 July 2018 - Added right side y tick marks label for D-01 to D-10

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Circle
%matplotlib inline
```

```python
# Data for extrapolating pathways for each descriptor is saved in the excel file
FILENAME = './data/pathways.xlsx'
# Standard figure size
FIG_SIZE = (12,8)
```

```python
"""
Functions for this notebook
"""
def circle(x, y, radius=1.2):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)

def text(x, y, text, bgcolor):
    ax.text(x, y, text, backgroundcolor=bgcolor,
            ha='center', va='top', weight='bold', color='k')

def circle2(x, y, radius=1.2):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax2.add_artist(circle)

def text2(x, y, text, bgcolor):
    ax2.text(x, y, text, backgroundcolor=bgcolor,
            ha='center', va='top', weight='bold', color='k')
    
def rightSide_ylabels():
    ax2 = ax.twinx()
    y1, y2 = ax.get_ylim()
    ax2.set_ylim(y1,y2)
    ax2.tick_params(axis='y', labelsize = 14)
```

```python
"""
Produce diagram for D-01 Population (Global).
Data was adapted from Schweizer and O'Niell (2014)
"""
Sheetname = 'Population'
Fig_y_label = 'Population in billions'
Fig_title = 'Global Population Pathways'

# Get data into Pandas dataframe
Popdf = pd.read_excel(FILENAME, sheet_name=Sheetname)

# Ploting the high and low pathways
ax = Popdf[['Year','LM','MH']].plot(x='Year',ylim=[5,15],figsize=FIG_SIZE, fontsize = 14, legend = False)
plt.title(Fig_title, fontsize=20)
ax.set_ylabel(Fig_y_label,fontsize=14)
ax.set_xlabel('Year',fontsize=14)

# Plot the shading for each region
ax.fill_between(Popdf['Year'], Popdf['LM'], Popdf['MH'],facecolor='aquamarine', interpolate=True)
ax.fill_between(Popdf['Year'], Popdf['MH'], Popdf['MH-region'],facecolor='khaki', interpolate=True)
ax.fill_between(Popdf['Year'], Popdf['LM-region'], Popdf['LM'],facecolor='lightblue', interpolate=True)

# Plot the text strings and labels
ax.text(.7, 0.75, 'High Pathways', rotation=21,transform=ax.transAxes, fontsize=14)
ax.text(.7, 0.49, 'Medium Pathways', rotation=0,transform=ax.transAxes, fontsize=14)
ax.text(.7, 0.32, 'Low Pathways', rotation=-6.5,transform=ax.transAxes, fontsize=14)

# Add right side y-label and tick marks
#ax2 = ax.twinx()
#y1, y2 = ax.get_ylim()
#ax2.set_ylim(y1,y2)
#ax2.tick_params(axis='y', labelsize = 14)

# Add right side y-label and tick marks
rightSide_ylabels()

plt.savefig('./figures/GPopulation.png') # Save diagram as .png
```
